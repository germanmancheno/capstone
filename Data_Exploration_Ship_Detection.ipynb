```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow, concatenate_images
```

```python
# Set paths
data_root = '../capstone/data/'
path_train = '../capstone/data/train_v2/'
masks = pd.read_csv(data_root + 'train_ship_segmentations_v2.csv')
```

# Structure of files

```python
curDir = os.getcwd()
print(curDir)
for i in os.listdir(curDir):
    print('   ' + i)
```

```python
# number of training files
path, dirs, files = next(os.walk(path_train))
print(path)
print(len(files))
```

# Plot some images with ships

```python
sampleSize = 4
imagesWithoutShips = masks[~masks['EncodedPixels'].isna()].sample(sampleSize)
fig, arr = plt.subplots(nrows=sampleSize,ncols=1, figsize=(40,40), constrained_layout=False)
for i, img in enumerate(imagesWithoutShips['ImageId']):
    print(i, img)
    row = i
    arr[row].imshow(imread(path_train + img))
```

# Plot some images without ships

```python
sampleSize = 4
imagesWithoutShips = masks[masks['EncodedPixels'].isna()].sample(sampleSize)
fig, arr = plt.subplots(nrows=sampleSize,ncols=1, figsize=(40,40), constrained_layout=False)
for i, img in enumerate(imagesWithoutShips['ImageId']):
    print(i, img)
    row = i
    #column = 1
    arr[row].imshow(imread(path_train + img))
```
