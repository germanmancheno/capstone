# Ship Detection

The objective of the project is the detection of ships in images. As a first step into this project, in this deliverable we will use 'decision trees' and 'random forests' to classify images that include ships. 

The steps that are taken are the following:
* Preprocessing of images: using only the R channel of each image, its values once transformed into a numpy array are discretized to either 0 or 255 setting a decision threshold in the value of 100
* Extraction of features: continuous jumps of values (from 0 to 255 and from 255 to 0) are seeked for throughout the image. In short, this is an approximation of how objects are detected, of which some geometrical characteristics are extracted. In this process, a discretization of the image is also made (for example, into a 2x2 grid - the grid size is parametrized)
* From a training set of images, a features vector for each image is created with 29 different features
* Creation of csv file for storing the features for 1000 training images
* Training a decision tree model and a random forest model
* Checking the efficiency of the tree model and the random forest model

The results obtained on this classification problem are summarized as follows:
* Decision tree accuracy: 73%
* Random forest accuracy: 80%

```python
# load and display an image with Matplotlib
import numpy as np
import pandas as pd
import math
from matplotlib import image
from matplotlib import pyplot

#listImages_WITH = ['00a9e2ec9.jpg', '00a52cd2a.jpg', '00abc623a.jpg', '00ce2c1c0.jpg', '00d0a646b.jpg', '00d412548.jpg', '00fd8e126.jpg', '0a0df8299.jpg']
#listImages_WITH = ['00e4b52b2.jpg', '0a1a58833.jpg']
#listImages_WITHOUT = ['000aed819.jpg', '000baef0c.jpg', '000c150b0.jpg', '000d42241.jpg', '000db3e62.jpg', '000e37fc6.jpg', '000e64855.jpg', '000f1f959.jpg', '000f7f2fd.jpg', '00a2a6f9e.jpg', '00a2dd872.jpg', '00a4ba88b.jpg']
#listImages_WITHOUT = ['000db3e62.jpg']
features_dict = {}
path_train = '../capstone/data/train_v2/'
data_root = '../capstone/data/'
masks = pd.read_csv(data_root + 'train_ship_segmentations_v2.csv')


list_of_images = []
#I reduced the number of images from 1000 to 20 for file sharing purposes. Actual training of decision tree and random forest has been made with 1000 images.
#for counter_image in range(1000):
for counter_image in range(20):
    image_ID = masks.iloc[counter_image][0]
    image_FLAG = masks.iloc[counter_image][1]
    print(image_ID,image_FLAG)
    #image1 = image.imread(path_train + '00ce2c1c0.jpg')
    #image1 = image.imread(path_train + imageWITH)
    image1 = image.imread(path_train + image_ID)
    color_avg_R = np.average(image1[:,:,0])
    color_avg_G = np.average(image1[:,:,1])
    color_avg_B = np.average(image1[:,:,2])
    if image_ID not in list_of_images:
        list_of_images.append(image_ID)
        if color_avg_R > 100 and color_avg_B > 100 and color_avg_G > 100:
            image0 = np.zeros(image1.shape)
            print('reflections')
            for i in range(768):
                for j in range(768):
                    if abs(image1[i,j,0]-color_avg_R) < 35:
                        image0[i,j,0] = 0
                    else:
                        image0[i,j,0] = 255
        else:
            # red discretized image
            image0 = np.zeros(image1.shape)
            for i in range(768):
                for j in range(768):
                    if image1[i,j,0] < 100:
                        image0[i,j,0] = 0
                    else:
                        image0[i,j,0] = 255

        #cleaning outliers
        delta = 5
        for k in range(3):
            delta = int(delta*(1-k*2/10))
            for i in range(delta,768-delta):
                for j in range(delta,768-delta):
                    flag1 = abs(image0[i,j,0]-image0[i-delta,j-delta,0]) > 50
                    flag2 = abs(image0[i,j,0]-image0[i-delta,j,0]) > 50
                    flag3 = abs(image0[i,j,0]-image0[i-delta,j+delta,0]) > 50
                    flag4 = abs(image0[i,j,0]-image0[i,j-delta,0]) > 50
                    flag5 = abs(image0[i,j,0]-image0[i,j+delta,0]) > 50
                    flag6 = abs(image0[i,j,0]-image0[i+delta,j-delta,0]) > 50
                    flag7 = abs(image0[i,j,0]-image0[i+delta,j,0]) > 50
                    flag8 = abs(image0[i,j,0]-image0[i+delta,j+delta,0]) > 50
                    counter = 0
                    if flag1:
                        counter += 1
                    if flag2:
                        counter += 1
                    if flag3:
                        counter += 1
                    if flag4:
                        counter += 1
                    if flag5:
                        counter += 1
                    if flag6:
                        counter += 1
                    if flag7:
                        counter += 1
                    if flag8:
                        counter += 1
                    if counter > 4:
                        image0[i,j,0] = image0[i,j-1,0]
        #black frame
        for j in range(768):
            for i in range(768):
                if i in range(20) or i in range(748,768) or j in range(20) or j in range(748,768):
                    image0[i,j,0] = 0
                    image0[i,j,1] = 0
                    image0[i,j,2] = 0


        #features 1 and 2
        #DETECTION OF HIGH RED COLOR GRADIENT
        #Vertical ships: slope_val = 0
        slope_val = 0 #not used yet
        feature1 = {}
        feature2 = {}
        feature1_detected = 0
        feature2_detected = 0
        number_of_pixels = 768
        number_of_divisions = 4 #1
        number_of_segments = int(np.sqrt(number_of_divisions))
        division_width = int(number_of_pixels/number_of_segments)

        init_pixel_x = 0
        final_pixel_x = division_width
        init_pixel_y = 0
        final_pixel_y = division_width

        for division_row in range(number_of_segments):
            for division_column in range(number_of_segments):
                feature1[(division_row+1,division_column+1)] = []
                feature2[(division_row+1,division_column+1)] = []
                init_pixel_row = division_row*division_width
                final_pixel_row = (division_row+1)*division_width
                init_pixel_column = division_column*division_width
                final_pixel_column = (division_column+1)*division_width

                for i in range(init_pixel_row,final_pixel_row):
                    feature1[(division_row+1,division_column+1)].append([0,0,0])
                    feature2[(division_row+1,division_column+1)].append([0,0,0])
                    feature1_detected = 0
                    feature2_detected = 0
                    for j in range(init_pixel_column+1,final_pixel_column):
                        if abs(image0[i,j,0] - image0[i,j-1,0]) > 50:
                            if feature1_detected == 0:
                                feature1_detected = 1
                                feature1[(division_row+1,division_column+1)][i-init_pixel_row] = [i,j,1]
                            elif feature2_detected == 0:
                                feature2_detected = 1
                                feature2[(division_row+1,division_column+1)][i-init_pixel_row] = [i,j,1]


        # feature 3
        feature3 = {}
        sum1 = {}
        sum2 = {}
        sum3 = {} #this is a reading of the deviation in "horizontal width"
        sum4 = {} #this is a reading of the deviation in "vertical width"

        for division_row in range(number_of_segments):
            for division_column in range(number_of_segments):
                feature3[(division_row+1,division_column+1)] = []
                sum1[(division_row+1,division_column+1)] = 0
                sum2[(division_row+1,division_column+1)] = 0
                init_pixel_row = division_row*division_width
                final_pixel_row = (division_row+1)*division_width
                init_pixel_column = division_column*division_width
                final_pixel_column = (division_column+1)*division_width
                for i in range(division_width):
                    feature3[(division_row+1,division_column+1)].append([0,0,0])
                    feature3[(division_row+1,division_column+1)][i][0] = feature1[(division_row+1,division_column+1)][i][0]
                    feature3[(division_row+1,division_column+1)][i][1] = feature1[(division_row+1,division_column+1)][i][1]
                    if (feature2[(division_row+1,division_column+1)][i][1]!=0 and feature1[(division_row+1,division_column+1)][i][1]!=0):# > 0:
                        feature3[(division_row+1,division_column+1)][i][2] = feature2[(division_row+1,division_column+1)][i][1] - feature1[(division_row+1,division_column+1)][i][1]
                        # sum1 and sum2 are used to calculate an average "horizontal width"
                        sum1[(division_row+1,division_column+1)] += 1
                        sum2[(division_row+1,division_column+1)] += feature3[(division_row+1,division_column+1)][i][2]


        # calculation of some features
        first = {}
        last = {}
        first_row = {}
        last_row = {}
        first_column = {}
        last_column = {}
        value1 = {}
        value2 = {}
        angle = {}
        length = {}
        width = {}
        roundness_factor = {}
        shipness_factor = {}
        delta = 10
        for division_row in range(number_of_segments):
            for division_column in range(number_of_segments):
                first[(division_row+1,division_column+1)] = 0
                last[(division_row+1,division_column+1)] = 0
                first_row[(division_row+1,division_column+1)] = 0
                last_row[(division_row+1,division_column+1)] = 0
                first_column[(division_row+1,division_column+1)] = 0
                last_column[(division_row+1,division_column+1)] = 0
                for i in range(delta,division_width-delta):
                    if feature3[(division_row+1,division_column+1)][i-10][2] == 0:
                        if feature3[(division_row+1,division_column+1)][i-9][2] == 0:
                            if feature3[(division_row+1,division_column+1)][i-8][2] == 0:
                                if feature3[(division_row+1,division_column+1)][i-7][2] == 0:
                                    if feature3[(division_row+1,division_column+1)][i-6][2] == 0:
                                        if feature3[(division_row+1,division_column+1)][i-5][2] == 0:
                                            if feature3[(division_row+1,division_column+1)][i-4][2] == 0:
                                                if feature3[(division_row+1,division_column+1)][i-3][2] == 0:
                                                    if feature3[(division_row+1,division_column+1)][i-2][2] == 0:
                                                        if feature3[(division_row+1,division_column+1)][i-1][2] == 0:
                                                            if feature3[(division_row+1,division_column+1)][i][2] != 0:
                                                                first[(division_row+1,division_column+1)] = feature3[(division_row+1,division_column+1)][i][0] #row
                                                                first_row[(division_row+1,division_column+1)] = first[(division_row+1,division_column+1)]
                                                                first_column[(division_row+1,division_column+1)] = feature3[(division_row+1,division_column+1)][i][1]
                    if feature3[(division_row+1,division_column+1)][i+10][2] == 0:
                        if feature3[(division_row+1,division_column+1)][i+9][2] == 0:
                            if feature3[(division_row+1,division_column+1)][i+8][2] == 0:
                                if feature3[(division_row+1,division_column+1)][i+7][2] == 0:
                                    if feature3[(division_row+1,division_column+1)][i+6][2] == 0:
                                        if feature3[(division_row+1,division_column+1)][i+5][2] == 0:
                                            if feature3[(division_row+1,division_column+1)][i+4][2] == 0:
                                                if feature3[(division_row+1,division_column+1)][i+3][2] == 0:
                                                    if feature3[(division_row+1,division_column+1)][i+2][2] == 0:
                                                        if feature3[(division_row+1,division_column+1)][i+1][2] == 0:
                                                            if feature3[(division_row+1,division_column+1)][i][2] != 0:
                                                                last[(division_row+1,division_column+1)] = feature3[(division_row+1,division_column+1)][i][0] #row
                                                                last_row[(division_row+1,division_column+1)] = last[(division_row+1,division_column+1)]
                                                                last_column[(division_row+1,division_column+1)] = feature3[(division_row+1,division_column+1)][i][1]

                #next parameters are numerator and denominator of the slope expression
                slope_numerator = (768-first_row[(division_row+1,division_column+1)])-(768-last_row[(division_row+1,division_column+1)])
                slope_denominator = first_column[(division_row+1,division_column+1)]-last_column[(division_row+1,division_column+1)]
                #next parameter need to be nonzero for an object to exist
                check_val1 = sum1[(division_row+1,division_column+1)]
                check_val2 = first_column[(division_row+1,division_column+1)]
                check_val3 = last_column[(division_row+1,division_column+1)]
    #            print("---------------------------------------------------------------------------------------")
    #            print("checks:",check_val1, check_val2, check_val3)
                if (check_val1 * check_val2 * check_val3 * slope_denominator != 0):
                    value1[(division_row+1,division_column+1)] = sum2[(division_row+1,division_column+1)]/sum1[(division_row+1,division_column+1)]
                    value2[(division_row+1,division_column+1)] = last[(division_row+1,division_column+1)] - first[(division_row+1,division_column+1)]
                    angle[(division_row+1,division_column+1)] = np.arctan(slope_numerator/slope_denominator)*180/np.pi

                    height_rows = sum2[(division_row+1,division_column+1)]/sum1[(division_row+1,division_column+1)]
                    width_columns = last_column[(division_row+1,division_column+1)] - first_column[(division_row+1,division_column+1)]
                    angle_scalar = np.arctan(slope_numerator/slope_denominator)
                    length[(division_row+1,division_column+1)] = abs(width_columns / np.cos(angle_scalar))
                    width[(division_row+1,division_column+1)] = abs(height_rows / np.cos(angle_scalar))
                else:
                    value1[(division_row+1,division_column+1)] = 0
                    value2[(division_row+1,division_column+1)] = 0
                    angle[(division_row+1,division_column+1)] = 999999 #this is a reading of cero objects in section
                    length[(division_row+1,division_column+1)] = 0
                    width[(division_row+1,division_column+1)] = 0

        #        print("---------------------------------------------------------------------------------------")
        #        print("value1 (width in a ship located horizontally): ",value1[(division_row+1,division_column+1)])
        #        print("value2 (length in a ship located vertically): ",value2[(division_row+1,division_column+1)])
        #        print("    first row: ", first_row[(division_row+1,division_column+1)])
        #        print("    first column: ", first_column[(division_row+1,division_column+1)])
        #        print("    last row: ", last_row[(division_row+1,division_column+1)])
        #        print("    last column: ", last_column[(division_row+1,division_column+1)])
        #        print("    length: ", length[(division_row+1,division_column+1)])
        #        print("    width: ", width[(division_row+1,division_column+1)])
                print("angle: ", angle[(division_row+1,division_column+1)], "deg")


        # number of sections with objects
        number_of_divisions_positive = 0
        for key, element in angle.items():
            if element != 999999:
                number_of_divisions_positive += 1
        print(number_of_divisions_positive)



        # definition of features vector
        features_vector = []
        #features_vector = [average_R, average_B, average_G, number of objects, 
                            # width1_horizontal, length1_horizontal, width1, length1, angle1,
                            # width2_horizontal, length2_horizontal, width2, length2, angle2, 
                            # width3_horizontal, length3_horizontal, width3, length3, angle4, 
                            # width4_horizontal, length4_horizontal, width4, length3, angle4, 
                            # 111111 (shipness) or 888888 (no shipness),
                            # 1 (there are ships) or 0 (there are no ships) ...]
        features_vector.append(color_avg_R)
        features_vector.append(color_avg_B)
        features_vector.append(color_avg_G)
        features_vector.append(number_of_divisions_positive)
        for division_row in range(number_of_segments):
            for division_column in range(number_of_segments):
                features_vector.append(value1[(division_row+1,division_column+1)])
                features_vector.append(value2[(division_row+1,division_column+1)])
                features_vector.append(length[(division_row+1,division_column+1)])
                features_vector.append(width[(division_row+1,division_column+1)])
                features_vector.append(angle[(division_row+1,division_column+1)])
                width_scalar = width[(division_row+1,division_column+1)]
                length_scalar = length[(division_row+1,division_column+1)]
                #initial measure of "shipness"
                if width_scalar != 0 and length_scalar/width_scalar > 2.5:
                    features_vector.append(111111)
                else:
                    features_vector.append(888888)

        # from training images, we determine whether images have ships or not
        if str(image_FLAG) == "nan":
            features_vector.append(0)
        else:
            features_vector.append(1)


        #    print(features_vector)
        features_dict[image_ID] = features_vector
        print("completed", image_ID)
        fig=pyplot.figure(figsize=(8, 8))
        fig.add_subplot(1, 2, 1)
        pyplot.imshow(image1)
        fig.add_subplot(1, 2, 2)
        pyplot.imshow(image0)
        pyplot.show()
    
print("\ncompleted")
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X = features_df.drop('there are ships', axis=1)
Y = features_df['there are ships']
```

```python
#but this is the one we really like since we use the dataframes that we already created with features and classes: X and Y 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
print("------------------")
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, Y_train)

# Performin some predictions
Y_test_pred_tree = clf.predict(X_test)
Y_train_pred_tree = clf.predict(X_train)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy on test set:",metrics.accuracy_score(Y_test, Y_test_pred_tree))
print("Accuracy on train set:",metrics.accuracy_score(Y_train, Y_train_pred_tree))
```
